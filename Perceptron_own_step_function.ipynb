```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
```

```python
X, Y = make_classification(n_samples=100, n_classes=2, n_features=2, n_informative=2, n_redundant=0, class_sep=1, random_state=1, n_clusters_per_class=1)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y, cmap='bwr', s=100, edgecolors='black')
```

We change the weights only when the point is missclassified

```python
def perceptron(X, Y):
  w1 = 1
  w2 = 1
  b = 1
  lr = 0.1
  epochs = 1000

  for j in range(epochs):

    for i in range(X.shape[0]):

      z = w1*X[i][0] + w2*X[i][1] + b

      if(z*Y[i] < 0):

        w1 = w1 + lr*Y[i]*X[i][0]
        w2 = w2 + lr*Y[i]*X[i][1]
        b = b + lr*Y[i]

  print(w1, w2, b)
```

```python
perceptron(X, Y)
```
